```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/t20i/t20i_info.csv')
import numpy as np

df
```

```python
# batting team
# bowling team
# city
# current s core
# ball left
# wickets left
# current rr
# last five overs
```

```python
# Pre-Processing

df.isnull().sum()
```

```python
# Replacing null values in city ( So we are trying to replace the venues in place of null cities )

cities=np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city']=cities
df['city'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.drop(columns='venue',inplace=True)
```

```python
df
```

```python
# Considering only few cities where balls throwed more than 600 i.e 5 matches
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df=df[df['city'].isin(eligible_cities)]
```

```python
df['current_score']=df.groupby('match_id')['runs'].cumsum()
df
```

```python
df['over']=df['ball'].apply(lambda x:str(x).split(".")[0])
df['balls']=df['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
df['balls_bowled']=(df['over'].astype('int')*6+df['balls'].astype('int'))
```

```python
df['balls_left']=120-df['balls_bowled']
df['balls_left']=df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df['players_dismissed']=df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['players_dismissed']=df['players_dismissed'].astype('int')
df['players_dismissed']=df.groupby('match_id')['players_dismissed'].cumsum()
df['wickets_left']=10-df['players_dismissed']
```

```python
df['crr']=(df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
# for getting something over a window can be done using pandas funcs
from pandas.tseries.offsets import LastWeekOfMonth
groups=df.groupby('match_id')
match_ids=df['match_id'].unique()
last_five=[]
for id in match_ids:
  last_five.extend(groups.get_group(id)['runs'].rolling(window=30).sum().values.tolist())
```

```python
# last_five overs runs for every match by window method to calculate it
```

```python
df['last_five']=last_five
df
```

```python
df=df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
df=df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df=df.sample(df.shape[0])
# Shuffling to avoid biased output randomizing rows
```

```python
df
```

```python
X=df.drop(columns=['runs_x'])
y=df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)  # 30k rows for training remaining for testing
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf=ColumnTransformer(
    [('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])],remainder='passthrough'
)

# One hot Encoding 'batting_team','bowling_team','city' attributes
```

```python
# from sklearn.compose import ColumnTransformer
# from sklearn.preprocessing import OneHotEncoder, StandardScaler
# from sklearn.pipeline import Pipeline
# from sklearn.ensemble import RandomForestRegressor

categorical_features = ['batting_team', 'bowling_team', 'city']
numerical_features = ['current_score', 'balls_left', 'wickets_left', 'crr', 'last_five']

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False, drop='first'), categorical_features),
        ('num', StandardScaler(), numerical_features)
    ],
    remainder='passthrough'
)

rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=1))
])

rf_pipeline.fit(X_train, y_train)
print("RandomForestRegressor pipeline created and trained successfully.")
```

```python
y_pred_rf = rf_pipeline.predict(X_test)

r2_rf = r2_score(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)

print(f"RandomForestRegressor R-squared: {r2_rf:.4f}")
print(f"RandomForestRegressor Mean Absolute Error: {mae_rf:.4f}")
```

```python
xgb_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', XGBRegressor(n_estimators=1000, random_state=1, learning_rate=0.2,max_depth=12))
])


xgb_pipeline.fit(X_train, y_train)
print("XGBRegressor pipeline created and trained successfully.")
```

```python
y_pred_xgb = xgb_pipeline.predict(X_test)

r2_xgb = r2_score(y_test, y_pred_xgb)
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)

print(f"XGBRegressor R-squared: {r2_xgb:.4f}")
print(f"XGBRegressor Mean Absolute Error: {mae_xgb:.4f}")
```

```python
print(f"\n--- Model Performance Comparison ---")
print(f"RandomForestRegressor R-squared: {r2_rf:.4f}")
print(f"RandomForestRegressor Mean Absolute Error: {mae_rf:.4f}")
print(f"XGBRegressor R-squared: {r2_xgb:.4f}")
print(f"XGBRegressor Mean Absolute Error: {mae_xgb:.4f}")

if r2_rf > r2_xgb and mae_rf < mae_xgb:
    print("\nBased on the R-squared and Mean Absolute Error, RandomForestRegressor performed better.")
elif r2_xgb > r2_rf and mae_xgb < mae_rf:
    print("\nBased on the R-squared and Mean Absolute Error, XGBRegressor performed better.")
else:
    print("\nThe performance of both models is comparable or requires further analysis to determine a clear winner.")
```
